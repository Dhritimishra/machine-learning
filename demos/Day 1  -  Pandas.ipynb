# Exercise: Dataframe operations - Find top 10 movies.

Dataset: https://github.com/abulbasar/data/tree/master/movielens

1. Load movies.csv as movies dataframe and ratings.csv as ratings dataframe
2. How many records are there in movies and ratings dataframe
3. What is the movie Id for movie Godfather? [Optional]
4. How many ratings the movie Godfather has received? [Optional]
5. Using ratings dataframe, calculate average rating and number of ratings for each movie
6. Remove all movie Id that have received fewer than 100 ratings.
7. Sort the data by avg rating and take top 10 records. 
8. Merge the output with movies dataframe to show title.


```python
movies = pd.read_csv("/data/movielens/movies.csv")
```

```python
ratings = pd.read_csv("/data/movielens/ratings.csv")
```

```python
movies.shape[0]
```

```python
ratings.shape[0]
```

```python
movies[movies.title.str.contains("Godfather")]
```

```python
ratings[ratings.movieId == 858].shape[0]
```

```python
ratings.groupby("movieId")\
.rating.agg(["mean", "count"])\
.query("count>100")\
.sort_values("mean", ascending = False)\
.iloc[:10, :]\
.merge(movies, on = "movieId")\
[["movieId", "mean", "title"]]
```

```python
ratings.groupby("movieId")\
.rating.agg(["mean", "count"])\
.query("count>100")\
.shape
```
