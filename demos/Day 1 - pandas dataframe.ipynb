```python
df.groupby("gender").charges.mean()
```

```python
df.groupby("gender").charges.agg(["mean", "median", "min", "max", "std"])
```

```python
df.groupby("gender").agg({"charges": "mean", "age": "mean"})
```

```python
df.groupby("gender").agg({"charges": ["mean", "min", "max"]
                          , "age": ["mean", "min", "max"]})
```

```python
df.groupby(["gender", "region"]).agg({"charges": ["mean", "min", "max"]
                          , "age": ["mean", "min", "max"]})
```

```python
df[(df.gender == "female") & (df.region == "northeast")].charges.mean()
```

```python
df.pivot_table("charges", "region", "gender", aggfunc="min")
```
