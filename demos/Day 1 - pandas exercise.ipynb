```python
import pandas as pd
```

```python
movies = pd.read_csv("/data/movielens/movies.csv")
ratings = pd.read_csv("/data/movielens/ratings.csv")
```

```python
godfather = movies[movies.title.str.contains("Godfather")]
godfather
```

```python
list(godfather.movieId)
```

```python
movieId = list(godfather.movieId)[0]
movieId
```

```python
ratings[ratings.movieId == movieId].shape
```

```python
agg = ratings.groupby("movieId").rating.agg(["mean", "count"])
agg
```

```python
agg.columns = ["avg", "len"]
agg
```

```python
agg = ratings.groupby("movieId").rating.agg(["mean", "count"])
agg = agg[agg["count"]>100]
#agg = agg[agg.count>100]
agg = agg.sort_values("mean", ascending = False)
agg = agg.iloc[:10, :]
agg
```

```python
type(agg.count)
```

```python
agg = ratings.groupby("movieId").rating.agg(["mean", "count"])
#agg = agg[agg["count"]>100]
agg = agg.query("count>100")
agg = agg.sort_values("mean", ascending = False)
agg = agg.iloc[:10, :]
agg = agg.merge(movies, on = "movieId")
agg = agg[["movieId", "mean", "title"]]
agg
```
