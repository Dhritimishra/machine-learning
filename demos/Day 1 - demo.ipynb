```python
import pandas as pd
```

```python
df = pd.read_csv("/data/bank-full.csv", sep = ";")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
y = label_encoder.fit_transform(df.y)
y
```

```python
X = df.drop(columns=["y"])
X.head()
```

```python
pd.options.display.max_columns = 1000
```

```python
X_dummy = pd.get_dummies(X)
X_dummy.sample(10)
```

```python
from sklearn import model_selection
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_dummy, y
                                                    , test_size = 0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(X_train)/len(X)
```

```python
from sklearn import linear_model
```

```python
est = linear_model.LogisticRegression()
```

```python
df.describe()
```

```python
scaling = preprocessing.StandardScaler()
X_train_std = scaling.fit_transform(X_train)
X_test_std = scaling.transform(X_test)
```

```python
pd.DataFrame(X_train_std).describe()
```

```python
est.fit(X_train_std, y_train)
```

```python
y_train_pred = est.predict(X_train_std)
y_test_pred = est.predict(X_test_std)
y_train_pred
```

```python
pd.DataFrame({"actual": y_train, "predicted": y_train_pred}).sample(10)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_train, y_train_pred)
```

```python
metrics.accuracy_score(y_test, y_test_pred)
```

```python
pd.Series(y_train).value_counts()/len(y_train)
```

```python
pd.Series(y_test).value_counts()/len(y_test)
```

```python
metrics.confusion_matrix(y_test, y_test_pred)
```

```python
y_test_prob = est.predict_proba(X_test_std)
y_test_prob.shape
```

```python
y_test_prob[:, 1]
```

```python
len(y_test_prob[:, 1])
```

```python
import numpy as np
```

```python
y_test_pred = np.where(y_test_prob[:, 1] > 0.5, 1, 0)
print(metrics.confusion_matrix(y_test, y_test_pred))
print(metrics.accuracy_score(y_test, y_test_pred))
```

```python
y_test_pred = np.where(y_test_prob[:, 1] > 0.8, 1, 0)
print(metrics.confusion_matrix(y_test, y_test_pred))
print(metrics.accuracy_score(y_test, y_test_pred))
```

```python
y_test_pred = np.where(y_test_prob[:, 1] > 0.1, 1, 0)
print(metrics.confusion_matrix(y_test, y_test_pred))
print(metrics.accuracy_score(y_test, y_test_pred))
```
